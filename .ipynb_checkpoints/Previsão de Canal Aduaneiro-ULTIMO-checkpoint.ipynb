```python
DATA_PATH   = 'sample_data.parquet'
CUTOFF_VAL  = pd.Timestamp('2024-07-01')
CUTOFF_TEST = pd.Timestamp('2024-11-01')

df = pd.read_parquet(DATA_PATH)
df = df.sort_values('registry_date').reset_index(drop=True)

# Quando registry_date for null, substituir por yearmonth
df['registry_date'] = df['registry_date'].fillna(df['yearmonth'])

# country_origin_code não deveria ser numerico, vamos transformar para string
df['country_origin_code'] = df['country_origin_code'].astype(str).str.replace(r'\.0$', '', regex=True)

global_ncm_mode = df['ncm_code'].mode().iloc[0]
df['ncm_code']  = df['ncm_code'].fillna(global_ncm_mode)

global_transp_mode      = df['transport_mode_pt'].mode().iloc[0]
df['transport_mode_pt'] = df['transport_mode_pt'].fillna(global_transp_mode)

mode_entry                  = df['clearance_place_entry'].mode().iloc[0]
df['clearance_place_entry'] = df['clearance_place_entry'].fillna(mode_entry)

mode_dispatch                  = df['clearance_place_dispatch'].mode().iloc[0]
df['clearance_place_dispatch'] = df['clearance_place_dispatch'].fillna(mode_dispatch)

mode_clearance        = df['clearance_place'].mode().iloc[0]
df['clearance_place'] = df['clearance_place'].fillna(mode_clearance)

global_shipper_mode = df['shipper_name'].mode().iloc[0]
df['shipper_name']  = df['shipper_name'].fillna(global_shipper_mode)

# Feature engineering
# Target Binário (0: Verde, 1: Outros)
df['target'] = (df['channel'] != 'VERDE').astype(int)
del df['channel']

# Features Temporais
df['mes'] = df['registry_date'].dt.month
df['dia_semana'] = df['registry_date'].dt.dayofweek

# NCM: Grau de Elaboração (Escala normalizada de 0 a 10)
df['ncm_grau_elaboracao'] = pd.to_numeric(df['ncm_code'].str[:2], errors='coerce')

# Categorização do tipo de importador e remetente por palavras-chave
keywords = [
    'INTERNATIONAL', 'TRADING', 'IMPORTADORA', 'INDUSTRIAL', 'COMERCIAL', 
    'EXPORTADORA', 'GLOBAL', 'CONTINENTAL', 'LOGÍSTICA', 'BRASIL', 
    'DISTRIBUIDORA', 'NACIONAL'
]
pattern = '(?i)(' + '|'.join(keywords) + ')'
df['consignee_tipo'] = df['consignee_name'].str.extract(pattern, expand=False).str.upper().fillna('OUTROS')
df['shipper_tipo']   = df['shipper_name'].str.extract(pattern, expand=False).str.upper().fillna('OUTROS')

del df['consignee_name']
del df['shipper_name']
del df['consignee_code']

# Identificação de Tipo de Recinto (Porto vs Aeroporto)
pattern_recinto = r'(?i)(PORTO|PORTUÁRIO|PORTUARIO|AEROPORTO)'
df['clearance_place_entry_tipo']    = df['clearance_place_entry'].str.extract(pattern_recinto, expand=False).str.upper().replace(['PORTUARIO', 'PORTUÁRIO'], 'PORTO').fillna('OUTROS')
df['clearance_place_dispatch_tipo'] = df['clearance_place_dispatch'].str.extract(pattern_recinto, expand=False).str.upper().replace(['PORTUARIO', 'PORTUÁRIO'], 'PORTO').fillna('OUTROS')
df['clearance_place_tipo']          = df['clearance_place'].str.extract(pattern_recinto, expand=False).str.upper().replace(['PORTUARIO', 'PORTUÁRIO'], 'PORTO').fillna('OUTROS')

# --- GET_DUMMIES ---
# Como a cardinalidade do ncm_code é baixa neste dataset, podemos usar dummies tranquilamente.
# Isso é exigido pelo scikit-learn (Random Forest) para lidar com variáveis textuais.
cols = ['clearance_place_dispatch', 'clearance_place_entry',
       'consignee_company_size', 'clearance_place', 'transport_mode_pt',
       'ncm_code', 'country_origin_code',  'consignee_tipo', 'shipper_tipo',
       'clearance_place_entry_tipo', 'clearance_place_dispatch_tipo',
       'clearance_place_tipo']

df_treated = pd.get_dummies(df, columns=cols)

cols_meta    = ['document_number', 'yearmonth', 'registry_date', 'target']
feature_cols = [c for c in df_treated.columns if c not in cols_meta]

# Split temporal
df_treated = df_treated.sort_values('registry_date').reset_index(drop=True)

CUTOFF_VAL  = pd.Timestamp('2024-07-01')
CUTOFF_TEST = pd.Timestamp('2024-11-01')

mask_train = df_treated['registry_date'] < CUTOFF_VAL
mask_val   = (df_treated['registry_date'] >= CUTOFF_VAL) & (df_treated['registry_date'] < CUTOFF_TEST)
mask_test  = df_treated['registry_date']  >= CUTOFF_TEST

X_train, y_train = df_treated.loc[mask_train, feature_cols], df_treated.loc[mask_train, 'target']
X_val,   y_val   = df_treated.loc[mask_val,   feature_cols], df_treated.loc[mask_val,   'target']
X_test,  y_test  = df_treated.loc[mask_test,  feature_cols], df_treated.loc[mask_test,  'target']

print(f"Features: {len(feature_cols)}")
print(f"Treino: {len(X_train):,}  |  Val: {len(X_val):,}  |  Teste: {len(X_test):,}")
print(f"Taxa Não-Verde — Treino: {y_train.mean():.2%}  Val: {y_val.mean():.2%}  Teste: {y_test.mean():.2%}")

# Baseline
# Sempre prever Verde
base_pred = np.zeros(len(y_test), dtype=int)
print(f"Previsão sempre verde → F1 Macro: {f1_score(y_test, base_pred, average='macro', zero_division=0):.4f}")

lr = LogisticRegression(class_weight='balanced', max_iter=500, solver='lbfgs', random_state=42)
lr.fit(X_train, y_train)
lr_proba = lr.predict_proba(X_test)[:, 1]
lr_pred  = lr.predict(X_test)
lr_auc = roc_auc_score(y_test, lr_proba)
lr_f1  = f1_score(y_test, lr_pred, average='macro')
print(f"Logistic Regression → AUC: {lr_auc:.4f}  F1 Macro: {lr_f1:.4f}")
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_sample_weight

# ---------------------------------------------------------
# 1. Carga e Tratamento Inicial
# ---------------------------------------------------------
DATA_PATH = 'sample_data.parquet'
CUTOFF_VAL = pd.Timestamp('2024-07-01')
CUTOFF_TEST = pd.Timestamp('2024-11-01')

df = pd.read_parquet(DATA_PATH)
df = df.sort_values('registry_date').reset_index(drop=True)

df['registry_date'] = df['registry_date'].fillna(df['yearmonth'])
df['country_origin_code'] = df['country_origin_code'].astype(str).str.replace(r'\.0$', '', regex=True)

# Preenchimento de Nulos pelas Modas
for col in ['ncm_code', 'transport_mode_pt', 'clearance_place_entry', 
            'clearance_place_dispatch', 'clearance_place', 'shipper_name']:
    df[col] = df[col].fillna(df[col].mode().iloc[0])

# ---------------------------------------------------------
# 2. Feature Engineering & Target
# ---------------------------------------------------------
# Target Etapa 1: Binário (0: Verde, 1: Outros)
df['target_s1'] = (df['channel'] != 'VERDE').astype(int)

df['mes'] = df['registry_date'].dt.month
df['dia_semana'] = df['registry_date'].dt.dayofweek
df['ncm_grau_elaboracao'] = pd.to_numeric(df['ncm_code'].str[:2], errors='coerce')

# Categorização de textos
keywords = ['INTERNATIONAL', 'TRADING', 'IMPORTADORA', 'INDUSTRIAL', 'COMERCIAL', 
            'EXPORTADORA', 'GLOBAL', 'CONTINENTAL', 'LOGÍSTICA', 'BRASIL', 
            'DISTRIBUIDORA', 'NACIONAL']
pattern = '(?i)(' + '|'.join(keywords) + ')'

df['consignee_tipo'] = df['consignee_name'].str.extract(pattern, expand=False).str.upper().fillna('OUTROS')
df['shipper_tipo'] = df['shipper_name'].str.extract(pattern, expand=False).str.upper().fillna('OUTROS')

df.drop(columns=['consignee_name', 'shipper_name', 'consignee_code'], inplace=True)

# Recintos
pattern_recinto = r'(?i)(PORTO|PORTUÁRIO|PORTUARIO|AEROPORTO)'
for col in ['clearance_place_entry', 'clearance_place_dispatch', 'clearance_place']:
    df[col+'_tipo'] = df[col].str.extract(pattern_recinto, expand=False).str.upper().replace(['PORTUARIO', 'PORTUÁRIO'], 'PORTO').fillna('OUTROS')

# ---------------------------------------------------------
# 3. Encoding e Split Temporal
# ---------------------------------------------------------
cols_to_dummy = ['clearance_place_dispatch', 'clearance_place_entry', 'consignee_company_size', 
                 'clearance_place', 'transport_mode_pt', 'ncm_code', 'country_origin_code', 
                 'consignee_tipo', 'shipper_tipo', 'clearance_place_entry_tipo', 
                 'clearance_place_dispatch_tipo', 'clearance_place_tipo']

df_treated = pd.get_dummies(df, columns=cols_to_dummy)

# XGBoost prefere nomes de colunas sem caracteres especiais (como <, >, [ ou ])
df_treated.columns = df_treated.columns.str.replace(r'[<\[\]]', '_', regex=True)

cols_meta = ['document_number', 'yearmonth', 'registry_date', 'target_s1', 'channel']
feature_cols = [c for c in df_treated.columns if c not in cols_meta]

df_treated = df_treated.sort_values('registry_date').reset_index(drop=True)

mask_train = df_treated['registry_date'] < CUTOFF_VAL
mask_val   = (df_treated['registry_date'] >= CUTOFF_VAL) & (df_treated['registry_date'] < CUTOFF_TEST)
mask_test  = df_treated['registry_date'] >= CUTOFF_TEST

X_train = df_treated.loc[mask_train, feature_cols]
X_val   = df_treated.loc[mask_val, feature_cols]
X_test  = df_treated.loc[mask_test, feature_cols]

# Targets
y_train_s1, y_train_s2 = df_treated.loc[mask_train, 'target_s1'], df_treated.loc[mask_train, 'channel']
y_test_s1,  y_test_s2  = df_treated.loc[mask_test, 'target_s1'],  df_treated.loc[mask_test, 'channel']

# ---------------------------------------------------------
# 4. Modelagem em Duas Etapas (Cascata com XGBoost)
# ---------------------------------------------------------

# --- ETAPA 1: Separar o Joio do Trigo (Verde vs Não-Verde) ---
print("Treinando Modelo Etapa 1 (XGBoost - Verde vs Outros)...")

# XGBoost lida com desbalanceamento muito melhor usando scale_pos_weight
# Calculamos a razão exata de Verdes para Riscos no conjunto de treino
peso_risco = (y_train_s1 == 0).sum() / (y_train_s1 == 1).sum()

# Instanciando e treinando XGBoost Binário com penalidade severa
xgb_s1 = xgb.XGBClassifier(
    scale_pos_weight=peso_risco, # A "mágica" para forçar o modelo a achar a classe minoritária
    random_state=42, 
    max_depth=6, 
    learning_rate=0.1, 
    n_estimators=150,
    eval_metric='logloss',
    n_jobs=-1
)
xgb_s1.fit(X_train, y_train_s1) # Agora treinamos com todo o dataset, o peso resolve o balanço


# --- ETAPA 2: Classificar o Risco (Amarelo, Vermelho, Cinza) ---
print("Treinando Modelo Etapa 2 (XGBoost - Apenas classes de risco)...")

mask_risk_train = y_train_s1 == 1
X_train_risk = X_train[mask_risk_train]
y_train_risk = y_train_s2[mask_risk_train]

# XGBoost multiclasse exige targets numéricos sequenciais (0, 1, 2)
risk_classes = ['AMARELO', 'VERMELHO', 'CINZA']
risk_mapping = {c: i for i, c in enumerate(risk_classes)}
risk_reverse = {i: c for i, c in enumerate(risk_classes)}

y_train_risk_num = y_train_risk.map(risk_mapping)

# Balanceamento através de pesos de amostra
sample_weights_s2 = compute_sample_weight(class_weight='balanced', y=y_train_risk_num)

xgb_s2 = xgb.XGBClassifier(
    objective='multi:softmax',
    num_class=len(risk_classes),
    random_state=42, 
    max_depth=6, 
    learning_rate=0.1,
    n_estimators=150,
    eval_metric='mlogloss',
    n_jobs=-1
)
xgb_s2.fit(X_train_risk, y_train_risk_num, sample_weight=sample_weights_s2)

# ---------------------------------------------------------
# 5. Inferência e Avaliação no Teste
# ---------------------------------------------------------
print("\nRealizando previsões no conjunto de teste...")

pred_test_s1 = xgb_s1.predict(X_test)

# --- AVALIAÇÃO EXCLUSIVA DA ETAPA 1 ---
print("\n=== Avaliação da Etapa 1 (Filtro Verde vs Risco) ===")
print("0: VERDE | 1: RISCO (Amarelo, Vermelho, Cinza)")
print(classification_report(y_test_s1, pred_test_s1))

cm_s1 = confusion_matrix(y_test_s1, pred_test_s1)
df_cm_s1 = pd.DataFrame(cm_s1, index=["Real VERDE (0)", "Real RISCO (1)"], columns=["Pred VERDE (0)", "Pred RISCO (1)"])
print("Matriz de Confusão (Etapa 1):")
print(df_cm_s1)
print("-" * 50)


# Todos começam como Verde por padrão
final_preds = np.array(['VERDE'] * len(X_test), dtype=object)

mask_pred_risk = pred_test_s1 == 1

# Se a ETAPA 1 disser que é risco (1), mandamos para a ETAPA 2
if mask_pred_risk.sum() > 0:
    X_test_risk = X_test[mask_pred_risk]
    
    # Previsão numérica (0, 1, 2)
    pred_test_s2_num = xgb_s2.predict(X_test_risk)
    
    # Revertendo para as strings originais ('AMARELO', 'VERMELHO', 'CINZA')
    pred_test_s2_str = np.vectorize(risk_reverse.get)(pred_test_s2_num)
    
    # Sobrescrevemos os "Verdes" apenas onde a Etapa 1 detectou risco
    final_preds[mask_pred_risk] = pred_test_s2_str

print("\n=== Resultado Final (Abordagem Cascata com XGBoost) ===")
print(classification_report(y_test_s2, final_preds))

print("Matriz de Confusão (Final):")
labels = ['VERDE', 'AMARELO', 'VERMELHO', 'CINZA']
present_labels = [l for l in labels if l in y_test_s2.unique() or l in np.unique(final_preds)]
cm = confusion_matrix(y_test_s2, final_preds, labels=present_labels)
print(pd.DataFrame(cm, index=[f"Real {l}" for l in present_labels], columns=[f"Pred {l}" for l in present_labels]))
```
