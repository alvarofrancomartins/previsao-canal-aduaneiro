```python
"""
=============================================================================
2.2 Pré-processamento dos Dados
=============================================================================
Continuação direta do notebook de EDA.
Premissas:
  - df já carregado com country_origin_code convertido para string
  - Nenhuma coluna temporária residual da EDA
"""

import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# 0. RELOAD (caso esteja rodando standalone — comentar se já tem df na memória)
# ============================================================================
# df = pd.read_parquet('sample_data.parquet')
# df['country_origin_code'] = df['country_origin_code'].astype(str).str.replace(r'\.0$', '', regex=True)
# df.loc[df['country_origin_code'] == 'nan', 'country_origin_code'] = pd.NA

# ============================================================================
# 1. TRATAMENTO DE DADOS FALTANTES
# ============================================================================
print("=" * 70)
print("1. TRATAMENTO DE DADOS FALTANTES")
print("=" * 70)

print(f"\nAntes do tratamento:\n{df.isnull().sum()[df.isnull().sum() > 0]}")

# 1a) registry_date (1300 nulls / 1.3%)
#     Estratégia: Preencher com o 1º dia do yearmonth correspondente.
#     Justificativa: yearmonth está 100% preenchido e é derivada de registry_date.
#     Mantemos a granularidade mensal sem inventar dia exato.
df['registry_date'] = df['registry_date'].fillna(df['yearmonth'])

# 1b) Categóricas com poucos nulls: clearance_place_entry (40), transport_mode_pt (40),
#     ncm_code (59), shipper_name (10)
#     Estratégia: Preencher com "DESCONHECIDO" — categoria explícita.
#     Justificativa: Volume < 0.06% cada. Imputar moda seria data leakage potencial
#     e apagaria o sinal de que o dado estava ausente (que pode ser informativo).
cols_cat_missing = ['clearance_place_entry', 'transport_mode_pt', 'ncm_code', 'shipper_name']
for col in cols_cat_missing:
    df[col] = df[col].fillna('DESCONHECIDO')

# 1c) country_origin_code (750 nulls / 0.75%)
#     Estratégia: Preencher com "DESCONHECIDO".
#     Justificativa: É uma variável de risco importante. Criar uma categoria
#     explícita preserva o sinal de ausência (que pode ser correlacionado com risco).
df['country_origin_code'] = df['country_origin_code'].fillna('DESCONHECIDO')

print(f"\nDepois do tratamento:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
print("✓ Zero nulls restantes." if df.isnull().sum().sum() == 0 else "⚠ Ainda há nulls!")

# ============================================================================
# 2. ENGENHARIA DE FEATURES
# ============================================================================
print("\n" + "=" * 70)
print("2. ENGENHARIA DE FEATURES")
print("=" * 70)

# Garantir ordenação cronológica (essencial para lag features)
df = df.sort_values('registry_date').reset_index(drop=True)

# --------------------------------------------------------------------------
# 2.1 Features Temporais
# --------------------------------------------------------------------------
print("\n[2.1] Features temporais...")

df['ano']          = df['registry_date'].dt.year
df['mes']          = df['registry_date'].dt.month
df['dia_semana']   = df['registry_date'].dt.dayofweek  # 0=Seg, 6=Dom
df['dia_mes']      = df['registry_date'].dt.day
df['trimestre']    = df['registry_date'].dt.quarter
df['semana_ano']   = df['registry_date'].dt.isocalendar().week.astype(int)
df['is_fim_semana'] = (df['dia_semana'] >= 5).astype(int)

# Flag: começo/fim de mês (primeiros/últimos 5 dias)
df['is_inicio_mes'] = (df['dia_mes'] <= 5).astype(int)
df['is_fim_mes']    = (df['dia_mes'] >= 25).astype(int)

# --------------------------------------------------------------------------
# 2.2 Captura de Sazonalidade — Encoding Cíclico (sin/cos)
# --------------------------------------------------------------------------
print("[2.2] Encoding cíclico para sazonalidade...")

# Mês: ciclo de 12
df['mes_sin'] = np.sin(2 * np.pi * df['mes'] / 12)
df['mes_cos'] = np.cos(2 * np.pi * df['mes'] / 12)

# Dia da semana: ciclo de 7
df['dia_semana_sin'] = np.sin(2 * np.pi * df['dia_semana'] / 7)
df['dia_semana_cos'] = np.cos(2 * np.pi * df['dia_semana'] / 7)

# Trimestre: ciclo de 4
df['trimestre_sin'] = np.sin(2 * np.pi * df['trimestre'] / 4)
df['trimestre_cos'] = np.cos(2 * np.pi * df['trimestre'] / 4)

# Semana do ano: ciclo de 52
df['semana_sin'] = np.sin(2 * np.pi * df['semana_ano'] / 52)
df['semana_cos'] = np.cos(2 * np.pi * df['semana_ano'] / 52)

# --------------------------------------------------------------------------
# 2.3 Features Derivadas da NCM
# --------------------------------------------------------------------------
print("[2.3] Features derivadas da NCM...")

# NCM tem 8 dígitos: os primeiros 2 são o capítulo, 4 é a posição, 6 é subposição
df['ncm_capitulo']   = df['ncm_code'].str[:2]   # Capítulo (ex: 39 = Plásticos)
df['ncm_posicao']    = df['ncm_code'].str[:4]   # Posição
df['ncm_subposicao'] = df['ncm_code'].str[:6]   # Subposição

# --------------------------------------------------------------------------
# 2.4 Features de Risco Históricas (Expanding Window — sem data leakage)
# --------------------------------------------------------------------------
print("[2.4] Features de risco com janela histórica (expanding window)...")

# Mapear target para numérico antes de calcular
risco_map = {'VERDE': 0, 'AMARELO': 1, 'VERMELHO': 2, 'CINZA': 3}
df['target_num'] = df['channel'].map(risco_map)

# Flag binária: é não-verde? (para calcular taxa de risco)
df['is_nao_verde'] = (df['channel'] != 'VERDE').astype(int)
df['is_vermelho']  = (df['channel'] == 'VERMELHO').astype(int)


def build_expanding_risk_features(df, group_col, prefix):
    """
    Calcula features de risco usando expanding window (tudo que veio ANTES).
    
    Para cada registro, usa apenas os dados históricos anteriores à data daquele 
    registro para calcular estatísticas de risco da categoria.
    
    Retorna:
      - {prefix}_vol:       volume acumulado de DIs da categoria
      - {prefix}_tx_risco:  taxa histórica de canal não-verde
      - {prefix}_tx_verm:   taxa histórica de canal vermelho
      - {prefix}_risco_med: risco médio histórico (escala 0-3)
    
    O shift(1) garante que não usamos o próprio registro no cálculo.
    """
    # Agrupamos por categoria e aplicamos expanding com shift(1)
    g = df.groupby(group_col)
    
    # Volume acumulado (count)
    df[f'{prefix}_vol'] = g['target_num'].cumcount()  # já exclui o próprio (conta de 0)
    
    # Taxa de risco (não-verde) acumulada
    df[f'{prefix}_tx_risco'] = (
        g['is_nao_verde']
        .expanding().sum().shift(1)
        .values
    )
    # Dividir pelo volume para ter a taxa
    df[f'{prefix}_tx_risco'] = np.where(
        df[f'{prefix}_vol'] > 0,
        df[f'{prefix}_tx_risco'] / df[f'{prefix}_vol'],
        0
    )
    
    # Taxa de vermelho acumulada
    df[f'{prefix}_tx_verm'] = (
        g['is_vermelho']
        .expanding().sum().shift(1)
        .values
    )
    df[f'{prefix}_tx_verm'] = np.where(
        df[f'{prefix}_vol'] > 0,
        df[f'{prefix}_tx_verm'] / df[f'{prefix}_vol'],
        0
    )
    
    # Risco médio acumulado
    df[f'{prefix}_risco_med'] = (
        g['target_num']
        .expanding().mean().shift(1)
        .values
    )
    df[f'{prefix}_risco_med'] = df[f'{prefix}_risco_med'].fillna(0)
    
    return df


# Aplicar para as entidades de risco mais relevantes
for col, prefix in [
    ('ncm_code',           'ncm'),
    ('ncm_capitulo',       'cap'),
    ('country_origin_code','pais'),
    ('consignee_code',     'imp'),       # importador
    ('shipper_name',       'ship'),
]:
    print(f"  → Calculando risco histórico para: {col} ({prefix}_*)")
    df = build_expanding_risk_features(df, col, prefix)

# --------------------------------------------------------------------------
# 2.5 Features de Risco por Combinação (interações)
# --------------------------------------------------------------------------
print("[2.5] Features de interação (NCM × País, Importador × NCM)...")

df['ncm_pais'] = df['ncm_code'] + '_' + df['country_origin_code']
df = build_expanding_risk_features(df, 'ncm_pais', 'ncmpais')

df['imp_ncm'] = df['consignee_code'] + '_' + df['ncm_code']
df = build_expanding_risk_features(df, 'imp_ncm', 'impncm')

# --------------------------------------------------------------------------
# 2.6 Lag Features — Volume e Risco dos Últimos Meses
# --------------------------------------------------------------------------
print("[2.6] Lag features mensais (volume e taxa de risco)...")

# Agrupar por yearmonth para calcular estatísticas mensais
monthly_stats = df.groupby('yearmonth').agg(
    vol_mensal     = ('target_num', 'count'),
    tx_risco_mes   = ('is_nao_verde', 'mean'),
    tx_verm_mes    = ('is_vermelho', 'mean'),
    risco_medio_mes= ('target_num', 'mean'),
).reset_index()

# Criar lags de 1, 2 e 3 meses
for lag in [1, 2, 3]:
    monthly_stats[f'vol_lag{lag}']        = monthly_stats['vol_mensal'].shift(lag)
    monthly_stats[f'tx_risco_lag{lag}']   = monthly_stats['tx_risco_mes'].shift(lag)
    monthly_stats[f'tx_verm_lag{lag}']    = monthly_stats['tx_verm_mes'].shift(lag)
    monthly_stats[f'risco_medio_lag{lag}']= monthly_stats['risco_medio_mes'].shift(lag)

# Média móvel de 3 meses (suavização)
monthly_stats['vol_mm3']      = monthly_stats['vol_mensal'].rolling(3).mean().shift(1)
monthly_stats['tx_risco_mm3'] = monthly_stats['tx_risco_mes'].rolling(3).mean().shift(1)
monthly_stats['tx_verm_mm3']  = monthly_stats['tx_verm_mes'].rolling(3).mean().shift(1)

# Dropar colunas do mês corrente (evitar leakage)
monthly_stats.drop(columns=['vol_mensal', 'tx_risco_mes', 'tx_verm_mes', 'risco_medio_mes'], inplace=True)

# Merge no df principal
df = df.merge(monthly_stats, on='yearmonth', how='left')

# Preencher NaN dos primeiros meses (sem histórico) com 0
lag_cols = [c for c in df.columns if 'lag' in c or '_mm3' in c]
df[lag_cols] = df[lag_cols].fillna(0)

# --------------------------------------------------------------------------
# 2.7 Limpeza de colunas auxiliares
# --------------------------------------------------------------------------
print("[2.7] Limpeza de colunas auxiliares...")

# Remover colunas intermediárias que não vão para o modelo
cols_drop_aux = ['is_nao_verde', 'is_vermelho', 'target_num', 'ncm_pais', 'imp_ncm']
df.drop(columns=cols_drop_aux, inplace=True, errors='ignore')

print(f"\n✓ Total de features após engenharia: {df.shape[1]} colunas")

# ============================================================================
# 3. ENCODING DE CATEGÓRICAS
# ============================================================================
print("\n" + "=" * 70)
print("3. ENCODING DE CATEGÓRICAS")
print("=" * 70)

# Separar tipos de categóricas por cardinalidade
cat_baixa_card = ['consignee_company_size', 'transport_mode_pt']  # ≤ 7 categorias
cat_alta_card  = [
    'clearance_place_dispatch', 'clearance_place_entry', 'clearance_place',  # 10-12 cats
    'ncm_capitulo', 'ncm_posicao', 'ncm_subposicao',
]
cat_muito_alta = ['consignee_code', 'consignee_name', 'ncm_code', 
                  'shipper_name', 'country_origin_code']

# 3a) Baixa cardinalidade → OrdinalEncoder (tree-friendly, mais eficiente que OHE)
print("\n[3a] Encoding ordinal para categóricas de baixa cardinalidade...")
oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
df[cat_baixa_card] = oe.fit_transform(df[cat_baixa_card].astype(str))

# 3b) Média cardinalidade → OrdinalEncoder (para tree-based models isso é suficiente)
print("[3b] Encoding ordinal para categóricas de média cardinalidade...")
oe2 = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
df[cat_alta_card] = oe2.fit_transform(df[cat_alta_card].astype(str))

# 3c) Alta cardinalidade → Já estão representadas pelas features de risco históricas
#     Dropar as colunas originais de texto para evitar leakage e ruído
#     As features de risco (imp_vol, imp_tx_risco, etc.) capturam a informação relevante
print("[3c] Dropando categóricas de alta cardinalidade (representadas por risk features)...")
cols_drop_high_card = ['consignee_name', 'shipper_name']  
# Mantemos consignee_code e ncm_code como ordinal para o modelo capturar padrões residuais
for col in ['consignee_code', 'ncm_code', 'country_origin_code']:
    df[col] = df[col].astype('category').cat.codes

df.drop(columns=cols_drop_high_card, inplace=True, errors='ignore')

# 3d) Target encoding
print("[3d] Encoding do target...")
le_target = LabelEncoder()
df['target'] = le_target.fit_transform(df['channel'])
print(f"    Classes: {dict(zip(le_target.classes_, le_target.transform(le_target.classes_)))}")

# ============================================================================
# 4. DIVISÃO TEMPORAL DOS DADOS
# ============================================================================
print("\n" + "=" * 70)
print("4. DIVISÃO TEMPORAL (TREINO / VALIDAÇÃO / TESTE)")
print("=" * 70)

# Identificar colunas que NÃO vão para o modelo
cols_meta = ['document_number', 'yearmonth', 'registry_date', 'channel']
feature_cols = [c for c in df.columns if c not in cols_meta + ['target']]

print(f"\nFeatures finais ({len(feature_cols)}):")
for i, c in enumerate(feature_cols, 1):
    print(f"  {i:2d}. {c}")

# Divisão temporal:
#   Treino:    Jan/2023 — Jun/2024  (~72% dos dados, 18 meses)
#   Validação: Jul/2024 — Out/2024  (~14% dos dados, 4 meses)
#   Teste:     Nov/2024 — Jan/2025  (~14% dos dados, 3 meses)
#
# Justificativa: Respeita a ordem cronológica (sem data leakage temporal).
# A validação inclui meses de vale (Jul) e pico (Out), testando robustez sazonal.
# O teste usa os dados mais recentes, simulando produção real.

cutoff_val  = pd.Timestamp('2024-07-01')
cutoff_test = pd.Timestamp('2024-11-01')

mask_train = df['registry_date'] < cutoff_val
mask_val   = (df['registry_date'] >= cutoff_val) & (df['registry_date'] < cutoff_test)
mask_test  = df['registry_date'] >= cutoff_test

X_train = df.loc[mask_train, feature_cols]
y_train = df.loc[mask_train, 'target']
X_val   = df.loc[mask_val, feature_cols]
y_val   = df.loc[mask_val, 'target']
X_test  = df.loc[mask_test, feature_cols]
y_test  = df.loc[mask_test, 'target']

print(f"\n{'Split':<12} {'Período':<25} {'Amostras':>10} {'% Total':>10}")
print("-" * 60)
print(f"{'Treino':<12} {'Jan/2023 — Jun/2024':<25} {len(X_train):>10,} {len(X_train)/len(df)*100:>9.1f}%")
print(f"{'Validação':<12} {'Jul/2024 — Out/2024':<25} {len(X_val):>10,} {len(X_val)/len(df)*100:>9.1f}%")
print(f"{'Teste':<12} {'Nov/2024 — Jan/2025':<25} {len(X_test):>10,} {len(X_test)/len(df)*100:>9.1f}%")

# Verificar distribuição do target em cada split
print("\nDistribuição do target por split:")
for nome, y in [('Treino', y_train), ('Validação', y_val), ('Teste', y_test)]:
    dist = y.value_counts(normalize=True).sort_index() * 100
    classes_str = " | ".join([f"{le_target.classes_[i]}: {v:.1f}%" for i, v in dist.items()])
    print(f"  {nome:<12} {classes_str}")

# ============================================================================
# 5. BALANCEAMENTO DE CLASSES
# ============================================================================
print("\n" + "=" * 70)
print("5. ESTRATÉGIAS DE BALANCEAMENTO")
print("=" * 70)

contagem_train = y_train.value_counts().sort_index()
for cls_idx in contagem_train.index:
    cls_name = le_target.classes_[cls_idx]
    n = contagem_train[cls_idx]
    print(f"  {cls_name:<12} {n:>6,} ({n/len(y_train)*100:.2f}%)")

print(f"""
Estratégia 1: class_weight='balanced' (nativo do modelo)
  → Penaliza erros em classes raras pelo inverso da frequência.
  → Sem overhead de memória, sem amostras sintéticas.
  → Implementação: parâmetro direto no LightGBM/XGBoost/RF.

Estratégia 2: sample_weight customizado (custo assimétrico por negócio)
  → Pesos calibrados pelo impacto de negócio de cada erro:
    - Errar CINZA (antifraude) é o mais caro → peso altíssimo
    - Errar VERMELHO é caro (inspeção física perdida) → peso alto
    - Errar AMARELO tem custo moderado → peso médio
    - Errar VERDE tem custo baixo (maioria) → peso base
  → Diferente de balanced: aqui o peso reflete risco de negócio,
    não apenas frequência estatística.
""")

# Substituir NaN/inf (segurança)
X_train = X_train.replace([np.inf, -np.inf], np.nan).fillna(0)
X_val   = X_val.replace([np.inf, -np.inf], np.nan).fillna(0)
X_test  = X_test.replace([np.inf, -np.inf], np.nan).fillna(0)

# Pré-calcular sample_weights para Estratégia 2
# Pesos baseados em custo de negócio (não apenas inverso da frequência)
custo_negocio = {
    le_target.transform(['VERDE'])[0]:    1.0,
    le_target.transform(['AMARELO'])[0]:  30.0,
    le_target.transform(['VERMELHO'])[0]: 50.0,
    le_target.transform(['CINZA'])[0]:    100.0,
}
sample_weights_train = y_train.map(custo_negocio).values

print("Sample weights (custo de negócio):")
for cls_idx, peso in custo_negocio.items():
    print(f"  {le_target.classes_[cls_idx]:<12} peso = {peso}")


# ============================================================================
# 6. RESUMO FINAL
# ============================================================================
print("\n" + "=" * 70)
print("6. RESUMO DO PRÉ-PROCESSAMENTO")
print("=" * 70)

print(f"""
Pipeline completo:
  ✓ Tratamento de nulls (registry_date via yearmonth, categóricas → 'DESCONHECIDO')
  ✓ Features temporais: ano, mês, dia_semana, trimestre, semana_ano, flags
  ✓ Encoding cíclico: sin/cos para mês, dia_semana, trimestre, semana_ano
  ✓ NCM derivadas: capítulo (2 dígitos), posição (4), subposição (6)
  ✓ Features de risco histórico (expanding window, sem leakage):
      - Por NCM, capítulo NCM, país, importador, shipper
      - Por interação NCM×País e Importador×NCM
      - Volume, taxa de risco, taxa de vermelho, risco médio
  ✓ Lag features mensais: volume e taxas de risco com lag 1/2/3 meses + MM3
  ✓ Encoding de categóricas (ordinal para tree-based)
  ✓ Divisão temporal: Treino → Val → Teste (cronológica, sem shuffle)
  ✓ Balanceamento: class_weight='balanced' + sample_weight por custo de negócio

Datasets disponíveis para modelagem:
  - X_train, y_train             (usar com class_weight ou sample_weights_train)
  - X_val, y_val                 (validação)
  - X_test, y_test               (teste final)
  - feature_cols                 (lista de features)
  - le_target                    (LabelEncoder para decode)
  - sample_weights_train         (pesos por custo de negócio)
  - custo_negocio                (dicionário classe → peso)
""")

# Verificação de sanidade
assert X_train.shape[1] == X_val.shape[1] == X_test.shape[1], "Mismatch no número de features!"
assert not X_train.isnull().any().any(), "NaN no treino!"
assert not X_val.isnull().any().any(), "NaN na validação!"
print("✓ Verificações de sanidade OK.")
```

```python
# --------------------------------------------------------------------------
# 0.1 Frequency Encoding para colunas de alta cardinalidade
# --------------------------------------------------------------------------
print("=" * 70)
print("0. ENCODING ADICIONAL (alta cardinalidade)")
print("=" * 70)

# Frequency encoding: substitui categoria pela sua frequência relativa no TREINO
# Vantagem: preserva informação de raridade sem explodir dimensionalidade
high_card_cols = ['consignee_name', 'shipper_name']

for col in high_card_cols:
    if col not in df.columns:
        print(f"  ⚠ {col} já foi dropada. Pulando.")
        continue
    
    freq_map = df.loc[mask_train, col].value_counts(normalize=True).to_dict()
    df[f'{col}_freq'] = df[col].map(freq_map).fillna(0)  # unseen → 0 (raro)
    print(f"  ✓ {col}_freq criada ({len(freq_map)} categorias)")

# --------------------------------------------------------------------------
# 0.2 Target Encoding com K-Fold (anti-leakage) para consignee_name
# --------------------------------------------------------------------------
print("\n[Target Encoding K-Fold] consignee_name e shipper_name...")

def target_encode_kfold(df, col, target_col, train_mask, n_splits=5, smoothing=20):
    """
    Target encoding com K-Fold no treino para evitar overfitting.
    Validação/teste usam a média global do treino.
    """
    col_encoded = f'{col}_te'
    df[col_encoded] = np.nan
    
    global_mean = df.loc[train_mask, target_col].mean()
    train_idx = df.index[train_mask]
    
    # K-Fold no treino
    kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    y_train_te = df.loc[train_mask, target_col]
    
    for fold_train, fold_val in kf.split(train_idx, y_train_te):
        fold_train_idx = train_idx[fold_train]
        fold_val_idx   = train_idx[fold_val]
        
        # Calcular média do target por categoria no fold de treino
        means = df.loc[fold_train_idx].groupby(col)[target_col].agg(['mean', 'count'])
        
        # Smoothing: categorias com poucos exemplos → puxar para média global
        smooth_means = (means['mean'] * means['count'] + global_mean * smoothing) / (means['count'] + smoothing)
        
        # Aplicar no fold de validação
        df.loc[fold_val_idx, col_encoded] = df.loc[fold_val_idx, col].map(smooth_means)
    
    # Para validação/teste: usar médias calculadas em TODO o treino
    train_means = df.loc[train_mask].groupby(col)[target_col].agg(['mean', 'count'])
    smooth_all  = (train_means['mean'] * train_means['count'] + global_mean * smoothing) / (train_means['count'] + smoothing)
    
    non_train_mask = ~train_mask
    df.loc[non_train_mask, col_encoded] = df.loc[non_train_mask, col].map(smooth_all)
    
    # Preencher unseen com média global
    df[col_encoded] = df[col_encoded].fillna(global_mean)
    
    return df

# Criar target binário para TE (is_nao_verde) — mais sinal que multiclass
df['_te_target'] = (df['channel'] != 'VERDE').astype(float)

for col in high_card_cols:
    if col not in df.columns:
        continue
    df = target_encode_kfold(df, col, '_te_target', mask_train)
    print(f"  ✓ {col}_te criada")

df.drop(columns=['_te_target'], inplace=True)

# Agora sim dropar as colunas texto (mantendo freq + te)
for col in high_card_cols:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

# --------------------------------------------------------------------------
# 0.3 Reconstruir feature_cols e splits
# --------------------------------------------------------------------------
cols_meta = ['document_number', 'yearmonth', 'registry_date', 'channel', 'target']
feature_cols = [c for c in df.columns if c not in cols_meta]

X_train = df.loc[mask_train, feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
y_train = df.loc[mask_train, 'target']
X_val   = df.loc[mask_val, feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
y_val   = df.loc[mask_val, 'target']
X_test  = df.loc[mask_test, feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
y_test  = df.loc[mask_test, 'target']

print(f"\n✓ Features finais: {len(feature_cols)}")


# ============================================================================
# 1. FUNÇÃO DE AVALIAÇÃO
# ============================================================================
results = {}

def evaluate_model(name, y_true, y_pred, y_pred_proba=None):
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, sup = precision_recall_fscore_support(
        y_true, y_pred, average=None, zero_division=0, labels=range(len(le_target.classes_))
    )
    f1_macro = f1_score(y_true, y_pred, average='macro', zero_division=0)
    f1_weighted = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    
    results[name] = {
        'accuracy': acc, 'f1_macro': f1_macro, 'f1_weighted': f1_weighted,
        'precision_per_class': prec, 'recall_per_class': rec,
        'f1_per_class': f1, 'support_per_class': sup,
        'y_pred': y_pred, 'y_pred_proba': y_pred_proba,
    }
    
    print(f"\n{'─' * 55}")
    print(f"  {name}")
    print(f"{'─' * 55}")
    print(f"  Acurácia: {acc:.4f}  |  F1 Macro: {f1_macro:.4f}  |  F1 Weighted: {f1_weighted:.4f}")
    print(f"\n  {'Classe':<12} {'Prec':>8} {'Recall':>8} {'F1':>8} {'N':>8}")
    print(f"  {'─'*44}")
    for i, cls_name in enumerate(le_target.classes_):
        flag = " ★" if cls_name != 'VERDE' and rec[i] > 0 else ""
        print(f"  {cls_name:<12} {prec[i]:>8.3f} {rec[i]:>8.3f} {f1[i]:>8.3f} {sup[i]:>8d}{flag}")
    
    return results[name]

# ============================================================================
# 2. MODELOS COM HIPERPARÂMETROS CORRIGIDOS
# ============================================================================
print("\n" + "=" * 70)
print("2. MODELOS PRINCIPAIS (hiperparâmetros corrigidos)")
print("=" * 70)

n_classes = len(le_target.classes_)

# Recalcular pesos balanceados
class_counts = np.bincount(y_train, minlength=n_classes)
balanced_weights = len(y_train) / (n_classes * class_counts)
lgb_sample_w = y_train.map(dict(enumerate(balanced_weights))).values

# --------------------------------------------------------------------------
# 2.1 LightGBM — balanced (CORRIGIDO)
# --------------------------------------------------------------------------
print("\n[2.1] LightGBM (balanced) — min_child_samples=5")

lgb_params = {
    'objective': 'multiclass',
    'num_class': n_classes,
    'metric': 'multi_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 63,
    'learning_rate': 0.03,          # mais conservador para não pular minorias
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'bagging_freq': 5,
    'min_child_samples': 5,         # ← FIX PRINCIPAL: era 50, CINZA tem 35
    'min_child_weight': 1e-3,       # ← sem restrição de peso mínimo
    'reg_alpha': 0.05,
    'reg_lambda': 0.5,
    'is_unbalance': True,           # ← LightGBM nativo para desbalanceamento
    'verbose': -1,
    'random_state': 42,
    'n_jobs': -1,
}

dtrain = lgb.Dataset(X_train, label=y_train, weight=lgb_sample_w)
dval   = lgb.Dataset(X_val, label=y_val, reference=dtrain)

t0 = time.time()
lgb_bal = lgb.train(
    lgb_params, dtrain, num_boost_round=3000,
    valid_sets=[dtrain, dval],
    callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)],
)
t1 = time.time() - t0

lgb_bal_proba = lgb_bal.predict(X_test)
lgb_bal_pred  = lgb_bal_proba.argmax(axis=1)
evaluate_model("LightGBM (balanced)", y_test, lgb_bal_pred, lgb_bal_proba)
print(f"  Treino: {t1:.1f}s | Iterações: {lgb_bal.best_iteration}")

# --------------------------------------------------------------------------
# 2.2 LightGBM — custo de negócio (CORRIGIDO)
# --------------------------------------------------------------------------
print("\n[2.2] LightGBM (custo de negócio)")

dtrain_sw = lgb.Dataset(X_train, label=y_train, weight=sample_weights_train)
dval_sw   = lgb.Dataset(X_val, label=y_val, reference=dtrain_sw)

lgb_params_custo = lgb_params.copy()
lgb_params_custo['is_unbalance'] = False  # pesos manuais, não automático

t0 = time.time()
lgb_custo = lgb.train(
    lgb_params_custo, dtrain_sw, num_boost_round=3000,
    valid_sets=[dtrain_sw, dval_sw],
    callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)],
)
t2 = time.time() - t0

lgb_custo_proba = lgb_custo.predict(X_test)
lgb_custo_pred  = lgb_custo_proba.argmax(axis=1)
evaluate_model("LightGBM (custo negócio)", y_test, lgb_custo_pred, lgb_custo_proba)
print(f"  Treino: {t2:.1f}s | Iterações: {lgb_custo.best_iteration}")

# --------------------------------------------------------------------------
# 2.3 XGBoost — balanced (CORRIGIDO)
# --------------------------------------------------------------------------
print("\n[2.3] XGBoost (balanced)")

xgb_params = {
    'objective': 'multi:softprob',
    'num_class': n_classes,
    'eval_metric': 'mlogloss',
    'max_depth': 7,
    'learning_rate': 0.03,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'min_child_weight': 1,          # ← FIX: era 50
    'gamma': 0.1,
    'reg_alpha': 0.05,
    'reg_lambda': 0.5,
    'tree_method': 'hist',
    'random_state': 42,
    'verbosity': 0,
}

dxgb_train = xgb.DMatrix(X_train, label=y_train, weight=lgb_sample_w)
dxgb_val   = xgb.DMatrix(X_val, label=y_val)
dxgb_test  = xgb.DMatrix(X_test, label=y_test)

t0 = time.time()
xgb_bal = xgb.train(
    xgb_params, dxgb_train, num_boost_round=3000,
    evals=[(dxgb_train, 'train'), (dxgb_val, 'val')],
    early_stopping_rounds=80, verbose_eval=False,
)
t3 = time.time() - t0

xgb_bal_proba = xgb_bal.predict(dxgb_test)
xgb_bal_pred  = xgb_bal_proba.argmax(axis=1)
evaluate_model("XGBoost (balanced)", y_test, xgb_bal_pred, xgb_bal_proba)
print(f"  Treino: {t3:.1f}s | Iterações: {xgb_bal.best_iteration}")

# --------------------------------------------------------------------------
# 2.4 XGBoost — custo de negócio (CORRIGIDO)
# --------------------------------------------------------------------------
print("\n[2.4] XGBoost (custo de negócio)")

dxgb_train_sw = xgb.DMatrix(X_train, label=y_train, weight=sample_weights_train)

t0 = time.time()
xgb_custo = xgb.train(
    xgb_params, dxgb_train_sw, num_boost_round=3000,
    evals=[(dxgb_train_sw, 'train'), (dxgb_val, 'val')],
    early_stopping_rounds=80, verbose_eval=False,
)
t4 = time.time() - t0

xgb_custo_proba = xgb_custo.predict(dxgb_test)
xgb_custo_pred  = xgb_custo_proba.argmax(axis=1)
evaluate_model("XGBoost (custo negócio)", y_test, xgb_custo_pred, xgb_custo_proba)
print(f"  Treino: {t4:.1f}s | Iterações: {xgb_custo.best_iteration}")

# --------------------------------------------------------------------------
# 2.5 Abordagem Hierárquica: Verde vs Não-Verde → Classificar Não-Verde
# --------------------------------------------------------------------------
print("\n[2.5] Modelo Hierárquico (2 estágios)")
print("  Estágio 1: Verde vs Não-Verde (binário)")
print("  Estágio 2: Classificar Não-Verde em Amarelo/Vermelho/Cinza")

# Estágio 1: Binário
y_train_bin = (y_train != le_target.transform(['VERDE'])[0]).astype(int)
y_val_bin   = (y_val   != le_target.transform(['VERDE'])[0]).astype(int)
y_test_bin  = (y_test  != le_target.transform(['VERDE'])[0]).astype(int)

# Peso muito agressivo para classe positiva (não-verde)
n_pos = y_train_bin.sum()
n_neg = len(y_train_bin) - n_pos
scale = n_neg / n_pos  # ~32x

lgb_bin_params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 63,
    'learning_rate': 0.03,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'bagging_freq': 5,
    'min_child_samples': 5,
    'scale_pos_weight': scale,     # peso proporcional ao desbalanceamento
    'verbose': -1,
    'random_state': 42,
}

dtrain_bin = lgb.Dataset(X_train, label=y_train_bin)
dval_bin   = lgb.Dataset(X_val, label=y_val_bin, reference=dtrain_bin)

lgb_stage1 = lgb.train(
    lgb_bin_params, dtrain_bin, num_boost_round=3000,
    valid_sets=[dtrain_bin, dval_bin],
    callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)],
)

stage1_proba = lgb_stage1.predict(X_test)

# Otimizar threshold no validation
stage1_proba_val = lgb_stage1.predict(X_val)
best_threshold, best_f1_bin = 0.5, 0
for t in np.arange(0.05, 0.95, 0.01):
    pred_bin = (stage1_proba_val >= t).astype(int)
    f1_bin = f1_score(y_val_bin, pred_bin, average='macro', zero_division=0)
    if f1_bin > best_f1_bin:
        best_f1_bin = f1_bin
        best_threshold = t

print(f"  Threshold ótimo (Estágio 1): {best_threshold:.2f} (F1 Macro val: {best_f1_bin:.4f})")

stage1_pred_test = (stage1_proba >= best_threshold).astype(int)
print(f"  Teste - Não-Verde detectados: {stage1_pred_test.sum()} de {y_test_bin.sum()} reais")
print(f"  Recall Não-Verde: {(stage1_pred_test[y_test_bin==1].sum() / max(y_test_bin.sum(),1)):.3f}")

# Estágio 2: Multiclass só nos Não-Verde
verde_idx = le_target.transform(['VERDE'])[0]
mask_nv_train = y_train != verde_idx

# Remapear classes: excluir VERDE do target
nv_classes = [c for c in range(n_classes) if c != verde_idx]
nv_remap   = {old: new for new, old in enumerate(nv_classes)}

X_train_nv = X_train[mask_nv_train]
y_train_nv = y_train[mask_nv_train].map(nv_remap)

# Pesos balanceados para o estágio 2
nv_counts = np.bincount(y_train_nv, minlength=len(nv_classes))
nv_weights = len(y_train_nv) / (len(nv_classes) * nv_counts + 1)
nv_sample_w = y_train_nv.map(dict(enumerate(nv_weights))).values

lgb_stage2_params = {
    'objective': 'multiclass',
    'num_class': len(nv_classes),
    'metric': 'multi_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.03,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'bagging_freq': 5,
    'min_child_samples': 3,       # ainda mais baixo — só 35 CINZA total
    'is_unbalance': True,
    'verbose': -1,
    'random_state': 42,
}

dtrain_nv = lgb.Dataset(X_train_nv, label=y_train_nv, weight=nv_sample_w)

# Validação do estágio 2
mask_nv_val = y_val != verde_idx
X_val_nv = X_val[mask_nv_val]
y_val_nv = y_val[mask_nv_val].map(nv_remap)
dval_nv  = lgb.Dataset(X_val_nv, label=y_val_nv, reference=dtrain_nv)

lgb_stage2 = lgb.train(
    lgb_stage2_params, dtrain_nv, num_boost_round=2000,
    valid_sets=[dtrain_nv, dval_nv],
    callbacks=[lgb.early_stopping(50), lgb.log_evaluation(0)],
)

# Combinar predições
nv_remap_inv = {v: k for k, v in nv_remap.items()}  # new → old
final_pred = np.full(len(y_test), verde_idx)

nv_mask_test = stage1_pred_test == 1
if nv_mask_test.sum() > 0:
    stage2_proba = lgb_stage2.predict(X_test[nv_mask_test])
    stage2_pred  = stage2_proba.argmax(axis=1)
    # Remapear de volta para labels originais
    final_pred[nv_mask_test] = np.array([nv_remap_inv[p] for p in stage2_pred])

evaluate_model("Hierárquico (2 estágios)", y_test, final_pred)
print(f"  Estágio 1 iters: {lgb_stage1.best_iteration} | Estágio 2 iters: {lgb_stage2.best_iteration}")


# ============================================================================
# 3. COMPARAÇÃO CONSOLIDADA
# ============================================================================
print("\n" + "=" * 70)
print("3. COMPARAÇÃO CONSOLIDADA")
print("=" * 70)

comparison = pd.DataFrame({
    name: {
        'Acurácia': r['accuracy'],
        'F1 Macro': r['f1_macro'],
        'F1 Weighted': r['f1_weighted'],
        'F1 VERDE': r['f1_per_class'][le_target.transform(['VERDE'])[0]],
        'F1 AMARELO': r['f1_per_class'][le_target.transform(['AMARELO'])[0]],
        'F1 VERMELHO': r['f1_per_class'][le_target.transform(['VERMELHO'])[0]],
        'F1 CINZA': r['f1_per_class'][le_target.transform(['CINZA'])[0]],
        'Recall VERM.': r['recall_per_class'][le_target.transform(['VERMELHO'])[0]],
        'Recall CINZA': r['recall_per_class'][le_target.transform(['CINZA'])[0]],
    }
    for name, r in results.items()
}).T.round(4)

print(f"\n{comparison.to_string()}")

best_model_name = comparison['F1 Macro'].idxmax()
print(f"\n★ Melhor modelo por F1 Macro: {best_model_name}")
print(f"  F1 Macro = {comparison.loc[best_model_name, 'F1 Macro']:.4f}")

# ============================================================================
# 4. MATRIZES DE CONFUSÃO
# ============================================================================
print("\n" + "=" * 70)
print("4. MATRIZES DE CONFUSÃO")
print("=" * 70)

n_models = len(results)
ncols = min(3, n_models)
nrows = (n_models + ncols - 1) // ncols
fig, axes = plt.subplots(nrows, ncols, figsize=(7*ncols, 6*nrows))
if n_models == 1:
    axes = np.array([axes])
axes = axes.ravel()

fig.suptitle('Matrizes de Confusão — Teste', fontsize=16, fontweight='bold', y=1.02)

for idx, (ax, name) in enumerate(zip(axes, results.keys())):
    cm = confusion_matrix(y_test, results[name]['y_pred'], labels=range(n_classes))
    cm_norm = cm.astype(float) / np.maximum(cm.sum(axis=1, keepdims=True), 1)
    
    annot = np.empty_like(cm, dtype=object)
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            annot[i, j] = f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)"
    
    sns.heatmap(
        cm_norm, annot=annot, fmt='', cmap='Blues', ax=ax,
        xticklabels=le_target.classes_, yticklabels=le_target.classes_,
        vmin=0, vmax=1, linewidths=0.5, linecolor='lightgrey',
    )
    ax.set_title(f'{name}\nF1M: {results[name]["f1_macro"]:.4f}', fontsize=11, fontweight='bold')
    ax.set_ylabel('Real')
    ax.set_xlabel('Predito')

# Esconder eixos extras
for ax in axes[len(results):]:
    ax.axis('off')

plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()

# ============================================================================
# 5. FEATURE IMPORTANCE (Melhor modelo LightGBM)
# ============================================================================
print("\n" + "=" * 70)
print("5. FEATURE IMPORTANCE")
print("=" * 70)

fig, axes = plt.subplots(1, 2, figsize=(18, 10))

# LightGBM balanced
lgb_imp = pd.Series(
    lgb_bal.feature_importance(importance_type='gain'), index=feature_cols
).sort_values(ascending=False).head(25)

sns.barplot(x=lgb_imp.values, y=lgb_imp.index, palette='viridis_r', edgecolor='k', ax=axes[0])
axes[0].set_title('LightGBM (balanced) — Gain Top 25', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Gain')
sns.despine(ax=axes[0])

# XGBoost balanced
xgb_imp_raw = xgb_bal.get_score(importance_type='gain')
xgb_imp = pd.Series(xgb_imp_raw).sort_values(ascending=False).head(25)

sns.barplot(x=xgb_imp.values, y=xgb_imp.index, palette='magma_r', edgecolor='k', ax=axes[1])
axes[1].set_title('XGBoost (balanced) — Gain Top 25', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Gain')
sns.despine(ax=axes[1])

plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nTop 10 Features (LightGBM):")
for i, (f, v) in enumerate(lgb_imp.head(10).items(), 1):
    print(f"  {i:2d}. {f:<35} gain = {v:,.0f}")

# ============================================================================
# 6. ROBUSTEZ SAZONAL
# ============================================================================
print("\n" + "=" * 70)
print("6. ROBUSTEZ SAZONAL")
print("=" * 70)

df_eval = pd.DataFrame({
    'y_true': y_test.values,
    'y_pred': results[best_model_name]['y_pred'],
    'yearmonth': df.loc[y_test.index, 'yearmonth'].values,
})

print(f"\n{best_model_name} por mês:\n")
print(f"  {'Mês':<10} {'N':>6} {'Acc':>8} {'F1M':>8} {'Rec.Verm':>10}")
print(f"  {'─'*44}")

verm_idx = le_target.transform(['VERMELHO'])[0]
for ym in sorted(df_eval['yearmonth'].unique()):
    s = df_eval[df_eval['yearmonth'] == ym]
    acc = accuracy_score(s['y_true'], s['y_pred'])
    f1m = f1_score(s['y_true'], s['y_pred'], average='macro', zero_division=0)
    # Recall vermelho no mês
    mask_v = s['y_true'] == verm_idx
    rec_v = (s.loc[mask_v, 'y_pred'] == verm_idx).mean() if mask_v.sum() > 0 else float('nan')
    print(f"  {pd.Timestamp(ym).strftime('%Y-%m'):<10} {len(s):>6} {acc:>8.4f} {f1m:>8.4f} {rec_v:>10.3f}")

# ============================================================================
# 7. RESPOSTAS MODELAGEM
# ============================================================================
print("\n" + "=" * 70)
print("7. RESPOSTAS — PERGUNTAS DE MODELAGEM")
print("=" * 70)

print(f"""
[5] Qual modelo apresentou melhor desempenho?

    {best_model_name} com F1 Macro = {comparison.loc[best_model_name, 'F1 Macro']:.4f}
    
    F1 Macro é a métrica primária: trata classes igualmente, não se esconde 
    atrás da acurácia inflada por VERDE (~97%). Recall de VERMELHO é métrica 
    secundária de negócio — cada não-detecção é carga de risco liberada.

[6] Estratégias de desbalanceamento (sem SMOTE):

    1. class_weight / is_unbalance: pesos inversamente proporcionais à 
       frequência. Simples, nativo dos frameworks, sem distorção dos dados.
    
    2. sample_weight por custo de negócio: pesos refletem impacto operacional.
       Liberar CINZA sem fiscalização antifraude ≫ fiscalizar VERDE à toa.
    
    3. Modelo hierárquico: separa decisão em 2 estágios (Verde vs Não-Verde,
       depois classifica o Não-Verde). Permite threshold customizado no 
       estágio 1 para controlar o trade-off precision/recall.
    
    Por que não SMOTE: CINZA com 35 amostras geraria pontos sintéticos 
    em regiões esparsas; features de expanding window não têm histórico 
    real para amostras sintéticas; viola estrutura cronológica.

[7] Features mais importantes → ver gráficos acima.
    Features de risco históricas (expanding window) tendem a dominar
    porque codificam o comportamento acumulado de cada entidade.

[8] Robustez sazonal:
    - Encoding cíclico sin/cos (mês, dia_semana, trimestre, semana)
    - Lag features 1-3 meses + média móvel 3 meses
    - Validação em meses de vale (Jul) E pico (Out)
    - Em produção: PSI monitoring + retreinamento trimestral com janela 12m
""")

# ============================================================================
# 8. SALVAR ARTEFATOS
# ============================================================================
print("=" * 70)
print("8. SALVANDO ARTEFATOS")
print("=" * 70)

lgb_bal.save_model('lgb_balanced.txt')
lgb_custo.save_model('lgb_custo_negocio.txt')
xgb_bal.save_model('xgb_balanced.json')
xgb_custo.save_model('xgb_custo_negocio.json')

metadata = {
    'feature_cols': feature_cols,
    'target_classes': list(le_target.classes_),
    'best_model': best_model_name,
    'metrics': {n: {'accuracy': float(r['accuracy']), 'f1_macro': float(r['f1_macro'])}
                for n, r in results.items()},
}
with open('model_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2, ensure_ascii=False)

print("  ✓ Modelos e metadata salvos.")
print("\n✓ Pipeline de modelagem v2 completo.")
```

```python
"""
=============================================================================
PIPELINE COMPLETO v4 — Previsão de Canal Aduaneiro
=============================================================================
Changelog v3→v4:
  - Baseline (Logistic Regression) incluído
  - country_origin_code incorporado na seção de Tratamento de Dados
  - Features de ROTA (entrada ≠ despacho ≠ desembaraço)
  - Flags de primeiro-visto por entidade (nova no sistema)
  - Features de combinação shipper×importador
  - Fix: hierarchical stage1 sem double-weighting
  - AUC-ROC por classe para avaliar ranking quality
  - Honestidade: diagnóstico claro da força do sinal
=============================================================================
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import lightgbm as lgb
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    classification_report, confusion_matrix, f1_score,
    accuracy_score, precision_recall_fscore_support, roc_auc_score
)
import time, json, warnings
warnings.filterwarnings('ignore')

# ============================================================================
# 0. CARGA
# ============================================================================
df = pd.read_parquet('sample_data.parquet')

# ============================================================================
# 1. TRATAMENTO DE DADOS
# ============================================================================
print("=" * 70)
print("1. TRATAMENTO DE DADOS")
print("=" * 70)

print(f"\nAntes:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
print(f"\nTipos:\n{df.dtypes}")

# 1a) country_origin_code: veio como float (ex: 160.0) → converter para string categórica
#     É um código de país, não um número contínuo. Tratar como categórica.
df['country_origin_code'] = (
    df['country_origin_code']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
)
df.loc[df['country_origin_code'] == 'nan', 'country_origin_code'] = pd.NA

# 1b) registry_date (1300 nulls / 1.3%)
#     yearmonth está 100% preenchido e é derivada de registry_date.
df['registry_date'] = df['registry_date'].fillna(df['yearmonth'])

# 1c) Categóricas com poucos nulls → 'DESCONHECIDO'
#     Preserva o sinal de ausência (dado faltante pode ser correlacionado com risco).
for col in ['clearance_place_entry', 'transport_mode_pt', 'ncm_code', 'shipper_name']:
    df[col] = df[col].fillna('DESCONHECIDO')

# 1d) country_origin_code (750 nulls)
df['country_origin_code'] = df['country_origin_code'].fillna('DESCONHECIDO')

print(f"\nDepois:\n{df.isnull().sum()[df.isnull().sum() > 0]}")
print("✓ Zero nulls." if df.isnull().sum().sum() == 0 else "⚠ Ainda há nulls!")

# ============================================================================
# 2. ENGENHARIA DE FEATURES
# ============================================================================
print("\n" + "=" * 70)
print("2. ENGENHARIA DE FEATURES")
print("=" * 70)

df = df.sort_values('registry_date').reset_index(drop=True)

# --- 2.1 Temporais ---
print("[2.1] Temporais...")
df['ano']           = df['registry_date'].dt.year
df['mes']           = df['registry_date'].dt.month
df['dia_semana']    = df['registry_date'].dt.dayofweek
df['dia_mes']       = df['registry_date'].dt.day
df['trimestre']     = df['registry_date'].dt.quarter
df['semana_ano']    = df['registry_date'].dt.isocalendar().week.astype(int)
df['is_fim_semana'] = (df['dia_semana'] >= 5).astype(int)
df['is_inicio_mes'] = (df['dia_mes'] <= 5).astype(int)
df['is_fim_mes']    = (df['dia_mes'] >= 25).astype(int)

# --- 2.2 Encoding cíclico ---
print("[2.2] Encoding cíclico...")
for col_name, period in [('mes', 12), ('dia_semana', 7), ('trimestre', 4), ('semana_ano', 52)]:
    df[f'{col_name}_sin'] = np.sin(2 * np.pi * df[col_name] / period)
    df[f'{col_name}_cos'] = np.cos(2 * np.pi * df[col_name] / period)

# --- 2.3 NCM derivadas ---
print("[2.3] NCM derivadas...")
df['ncm_capitulo']   = df['ncm_code'].str[:2]
df['ncm_posicao']    = df['ncm_code'].str[:4]
df['ncm_subposicao'] = df['ncm_code'].str[:6]

# --- 2.4 Features de ROTA (dicionário de dados!) ---
print("[2.4] Features de rota...")
# Mercadoria que ENTRA em um lugar e é DESPACHADA em outro → rota complexa
# Rota complexa pode indicar operação atípica (mais risco)
df['rota_entry_dispatch_diff'] = (
    df['clearance_place_entry'] != df['clearance_place_dispatch']
).astype(int)
df['rota_entry_place_diff'] = (
    df['clearance_place_entry'] != df['clearance_place']
).astype(int)
df['rota_dispatch_place_diff'] = (
    df['clearance_place_dispatch'] != df['clearance_place']
).astype(int)
# Nível de complexidade da rota (0=tudo mesmo local, 3=tudo diferente)
df['rota_complexidade'] = (
    df['rota_entry_dispatch_diff'] +
    df['rota_entry_place_diff'] +
    df['rota_dispatch_place_diff']
)
print(f"  Distribuição rota_complexidade: {df['rota_complexidade'].value_counts().sort_index().to_dict()}")

# --- 2.5 Features de risco históricas (expanding, CORRIGIDAS) ---
print("[2.5] Risco histórico (expanding window corrigida)...")

risco_map = {'VERDE': 0, 'AMARELO': 1, 'VERMELHO': 2, 'CINZA': 3}
df['target_num']   = df['channel'].map(risco_map)
df['is_nao_verde'] = (df['channel'] != 'VERDE').astype(int)
df['is_vermelho']  = (df['channel'] == 'VERMELHO').astype(int)


def build_risk_features(df, group_col, prefix):
    """Expanding window COM shift por grupo (sem leakage entre grupos)."""
    g = df.groupby(group_col)
    
    df[f'{prefix}_vol'] = g['target_num'].cumcount()
    
    df[f'{prefix}_sum_risco'] = g['is_nao_verde'].transform(
        lambda x: x.expanding().sum().shift(1)
    )
    df[f'{prefix}_tx_risco'] = np.where(
        df[f'{prefix}_vol'] > 0,
        df[f'{prefix}_sum_risco'] / df[f'{prefix}_vol'],
        0
    )
    
    df[f'{prefix}_sum_verm'] = g['is_vermelho'].transform(
        lambda x: x.expanding().sum().shift(1)
    )
    df[f'{prefix}_tx_verm'] = np.where(
        df[f'{prefix}_vol'] > 0,
        df[f'{prefix}_sum_verm'] / df[f'{prefix}_vol'],
        0
    )
    
    df[f'{prefix}_risco_med'] = g['target_num'].transform(
        lambda x: x.expanding().mean().shift(1)
    ).fillna(0)
    
    # FLAG: Primeira vez que vemos esta entidade (volume = 0)
    df[f'{prefix}_first_seen'] = (df[f'{prefix}_vol'] == 0).astype(int)
    
    df.drop(columns=[f'{prefix}_sum_risco', f'{prefix}_sum_verm'], inplace=True)
    return df


for col, prefix in [
    ('ncm_code',            'ncm'),
    ('ncm_capitulo',        'cap'),
    ('country_origin_code', 'pais'),
    ('consignee_code',      'imp'),
    ('shipper_name',        'ship'),
]:
    print(f"  → {col} ({prefix}_*)")
    df = build_risk_features(df, col, prefix)

# --- 2.6 Interações ---
print("[2.6] Interações...")
df['ncm_pais'] = df['ncm_code'] + '_' + df['country_origin_code']
df = build_risk_features(df, 'ncm_pais', 'ncmpais')

df['imp_ncm'] = df['consignee_code'] + '_' + df['ncm_code']
df = build_risk_features(df, 'imp_ncm', 'impncm')

# Shipper × Importador (relação comercial)
df['ship_imp'] = df['shipper_name'] + '_' + df['consignee_code']
df = build_risk_features(df, 'ship_imp', 'shipimp')

# --- 2.7 Lags mensais ---
print("[2.7] Lags mensais...")
monthly = df.groupby('yearmonth').agg(
    vol=('target_num', 'count'),
    tx_risco=('is_nao_verde', 'mean'),
    tx_verm=('is_vermelho', 'mean'),
).reset_index()

for lag in [1, 2, 3]:
    monthly[f'vol_lag{lag}']      = monthly['vol'].shift(lag)
    monthly[f'tx_risco_lag{lag}'] = monthly['tx_risco'].shift(lag)
    monthly[f'tx_verm_lag{lag}']  = monthly['tx_verm'].shift(lag)

monthly['vol_mm3']      = monthly['vol'].rolling(3).mean().shift(1)
monthly['tx_risco_mm3'] = monthly['tx_risco'].rolling(3).mean().shift(1)
monthly['tx_verm_mm3']  = monthly['tx_verm'].rolling(3).mean().shift(1)

monthly.drop(columns=['vol', 'tx_risco', 'tx_verm'], inplace=True)
df = df.merge(monthly, on='yearmonth', how='left')

lag_cols = [c for c in df.columns if 'lag' in c or '_mm3' in c]
df[lag_cols] = df[lag_cols].fillna(0)

# --- Limpeza ---
df.drop(columns=['is_nao_verde', 'is_vermelho', 'target_num',
                  'ncm_pais', 'imp_ncm', 'ship_imp'], inplace=True, errors='ignore')

print(f"\n✓ Colunas: {df.shape[1]}")

# ============================================================================
# 3. ENCODING
# ============================================================================
print("\n" + "=" * 70)
print("3. ENCODING")
print("=" * 70)

# Target
le_target = LabelEncoder()
df['target'] = le_target.fit_transform(df['channel'])
print(f"Classes: {dict(zip(le_target.classes_, le_target.transform(le_target.classes_)))}")

# Split temporal (ANTES do target encoding para evitar leakage)
cutoff_val  = pd.Timestamp('2024-07-01')
cutoff_test = pd.Timestamp('2024-11-01')
mask_train = df['registry_date'] < cutoff_val
mask_val   = (df['registry_date'] >= cutoff_val) & (df['registry_date'] < cutoff_test)
mask_test  = df['registry_date'] >= cutoff_test

# 3a) Baixa/média cardinalidade → Ordinal
cat_baixa = ['consignee_company_size', 'transport_mode_pt']
cat_media = ['clearance_place_dispatch', 'clearance_place_entry', 'clearance_place',
             'ncm_capitulo', 'ncm_posicao', 'ncm_subposicao']

oe1 = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
df[cat_baixa] = oe1.fit_transform(df[cat_baixa].astype(str))
oe2 = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
df[cat_media] = oe2.fit_transform(df[cat_media].astype(str))

# 3b) Alta cardinalidade → Freq + Target Encoding
high_card = ['consignee_code', 'ncm_code', 'country_origin_code', 'consignee_name', 'shipper_name']

print("\n[Frequency Encoding]")
for col in high_card:
    freq_map = df.loc[mask_train, col].value_counts(normalize=True).to_dict()
    df[f'{col}_freq'] = df[col].map(freq_map).fillna(0)
    print(f"  ✓ {col}_freq")

print("\n[Target Encoding K-Fold]")
df['_te_target'] = (df['channel'] != 'VERDE').astype(float)

def target_encode_kfold(df, col, target_col, train_mask, n_splits=5, smoothing=20):
    col_enc = f'{col}_te'
    df[col_enc] = np.nan
    global_mean = df.loc[train_mask, target_col].mean()
    train_idx = df.index[train_mask]
    y_te = df.loc[train_mask, target_col]
    
    kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    for ft, fv in kf.split(train_idx, y_te):
        ft_idx, fv_idx = train_idx[ft], train_idx[fv]
        stats = df.loc[ft_idx].groupby(col)[target_col].agg(['mean', 'count'])
        smooth = (stats['mean'] * stats['count'] + global_mean * smoothing) / (stats['count'] + smoothing)
        df.loc[fv_idx, col_enc] = df.loc[fv_idx, col].map(smooth)
    
    stats_all = df.loc[train_mask].groupby(col)[target_col].agg(['mean', 'count'])
    smooth_all = (stats_all['mean'] * stats_all['count'] + global_mean * smoothing) / (stats_all['count'] + smoothing)
    df.loc[~train_mask, col_enc] = df.loc[~train_mask, col].map(smooth_all)
    df[col_enc] = df[col_enc].fillna(global_mean)
    return df

for col in high_card:
    df = target_encode_kfold(df, col, '_te_target', mask_train)
    print(f"  ✓ {col}_te")

df.drop(columns=['_te_target'], inplace=True)
df.drop(columns=high_card, inplace=True)

# ============================================================================
# 4. SPLIT FINAL
# ============================================================================
print("\n" + "=" * 70)
print("4. SPLIT TEMPORAL")
print("=" * 70)

cols_meta = ['document_number', 'yearmonth', 'registry_date', 'channel', 'target']
feature_cols = [c for c in df.columns if c not in cols_meta]

X_train = df.loc[mask_train, feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
y_train = df.loc[mask_train, 'target']
X_val   = df.loc[mask_val, feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
y_val   = df.loc[mask_val, 'target']
X_test  = df.loc[mask_test, feature_cols].replace([np.inf, -np.inf], np.nan).fillna(0)
y_test  = df.loc[mask_test, 'target']

n_classes = len(le_target.classes_)
verde_idx = le_target.transform(['VERDE'])[0]

print(f"Features: {len(feature_cols)}")
print(f"Treino: {len(X_train):,} | Val: {len(X_val):,} | Teste: {len(X_test):,}")

for nome, y in [('Treino', y_train), ('Val', y_val), ('Teste', y_test)]:
    dist = y.value_counts(normalize=True).sort_index() * 100
    s = " | ".join([f"{le_target.classes_[i]}: {v:.1f}%" for i, v in dist.items()])
    print(f"  {nome:<8} {s}")

# Sample weights
class_counts = np.bincount(y_train, minlength=n_classes)
balanced_weights = len(y_train) / (n_classes * class_counts + 1)
sample_w_balanced = y_train.map(dict(enumerate(balanced_weights))).values

custo_negocio = {
    le_target.transform(['VERDE'])[0]:    1.0,
    le_target.transform(['AMARELO'])[0]:  30.0,
    le_target.transform(['VERMELHO'])[0]: 50.0,
    le_target.transform(['CINZA'])[0]:    100.0,
}
sample_w_custo = y_train.map(custo_negocio).values

# ============================================================================
# 5. UTILIDADES
# ============================================================================
results = {}

def evaluate_model(name, y_true, y_pred, y_pred_proba=None):
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, sup = precision_recall_fscore_support(
        y_true, y_pred, average=None, zero_division=0, labels=range(n_classes)
    )
    f1_macro = f1_score(y_true, y_pred, average='macro', zero_division=0)
    
    # AUC-ROC (One-vs-Rest) se temos probabilidades
    auc_ovr = None
    if y_pred_proba is not None and len(np.unique(y_true)) > 1:
        try:
            auc_ovr = roc_auc_score(y_true, y_pred_proba, multi_class='ovr', average='macro')
        except:
            auc_ovr = None
    
    results[name] = {
        'accuracy': acc, 'f1_macro': f1_macro, 'auc_ovr': auc_ovr,
        'f1_weighted': f1_score(y_true, y_pred, average='weighted', zero_division=0),
        'precision_per_class': prec, 'recall_per_class': rec,
        'f1_per_class': f1, 'support_per_class': sup,
        'y_pred': y_pred, 'y_pred_proba': y_pred_proba,
    }
    
    print(f"\n{'─'*60}")
    print(f"  {name}")
    print(f"{'─'*60}")
    auc_str = f" | AUC-ROC: {auc_ovr:.4f}" if auc_ovr else ""
    print(f"  Acc: {acc:.4f} | F1 Macro: {f1_macro:.4f}{auc_str}")
    print(f"\n  {'Classe':<12} {'Prec':>8} {'Recall':>8} {'F1':>8} {'N':>8}")
    print(f"  {'─'*44}")
    for i, cn in enumerate(le_target.classes_):
        flag = " ★" if cn != 'VERDE' and rec[i] > 0 else ""
        print(f"  {cn:<12} {prec[i]:>8.3f} {rec[i]:>8.3f} {f1[i]:>8.3f} {sup[i]:>8d}{flag}")


def optimize_thresholds_ovr(y_true, proba, class_names):
    """Otimiza threshold por classe (One-vs-Rest) na validação."""
    thresholds = np.full(len(class_names), 0.5)
    for i, cn in enumerate(class_names):
        if cn == 'VERDE':
            continue
        y_bin = (y_true == i).astype(int)
        proba_i = proba[:, i]
        
        best_f1, best_t = 0, 0.5
        for t in np.arange(0.001, 0.30, 0.001):
            pred = (proba_i >= t).astype(int)
            if pred.sum() == 0:
                continue
            f1 = f1_score(y_bin, pred, zero_division=0)
            if f1 > best_f1:
                best_f1 = f1
                best_t = t
        thresholds[i] = best_t
        print(f"  {cn:<12} t={best_t:.4f} F1_bin={best_f1:.4f} N_pred={int((proba_i>=best_t).sum())}")
    return thresholds


def predict_with_thresholds(proba, thresholds, verde_idx):
    """Predição com thresholds: entre as classes que passam, escolhe a de maior P."""
    preds = np.full(len(proba), verde_idx)
    for i in range(len(proba)):
        above = [(proba[i, c], c) for c in range(proba.shape[1])
                 if c != verde_idx and proba[i, c] >= thresholds[c]]
        if above:
            above.sort(reverse=True)
            preds[i] = above[0][1]
    return preds


# ============================================================================
# 6. BASELINES
# ============================================================================
print("\n" + "=" * 70)
print("6. BASELINES")
print("=" * 70)

# Baseline ingênuo: sempre VERDE
baseline_pred = np.full(len(y_test), verde_idx)
evaluate_model("Baseline (sempre VERDE)", y_test, baseline_pred)

# Baseline linear: Logistic Regression
print("\n  Treinando Logistic Regression...")
lr = LogisticRegression(
    class_weight='balanced', max_iter=1000, random_state=42,
    solver='lbfgs', multi_class='multinomial', C=0.1
)
lr.fit(X_train, y_train)
lr_pred  = lr.predict(X_test)
lr_proba = lr.predict_proba(X_test)

# LR com threshold
lr_val_proba = lr.predict_proba(X_val)
print("\n  Thresholds LR (validação):")
thresholds_lr = optimize_thresholds_ovr(y_val.values, lr_val_proba, le_target.classes_)
lr_pred_tuned = predict_with_thresholds(lr_proba, thresholds_lr, verde_idx)

evaluate_model("Logistic Regression (argmax)", y_test, lr_pred, lr_proba)
evaluate_model("Logistic Regression (threshold)", y_test, lr_pred_tuned, lr_proba)

# ============================================================================
# 7. MODELOS PRINCIPAIS
# ============================================================================
print("\n" + "=" * 70)
print("7. MODELOS PRINCIPAIS")
print("=" * 70)

lgb_params = {
    'objective': 'multiclass',
    'num_class': n_classes,
    'metric': 'multi_logloss',
    'boosting_type': 'gbdt',
    'num_leaves': 63,
    'learning_rate': 0.03,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'bagging_freq': 5,
    'min_child_samples': 5,
    'min_child_weight': 1e-3,
    'is_unbalance': True,
    'verbose': -1,
    'random_state': 42,
    'n_jobs': -1,
}

# --- 7.1 LightGBM balanced ---
print("\n[7.1] LightGBM (balanced)...")
dtrain = lgb.Dataset(X_train, label=y_train, weight=sample_w_balanced)
dval   = lgb.Dataset(X_val, label=y_val, reference=dtrain)

lgb_bal = lgb.train(
    lgb_params, dtrain, num_boost_round=3000,
    valid_sets=[dtrain, dval],
    callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)],
)
print(f"  Iterações: {lgb_bal.best_iteration}")

lgb_val_proba  = lgb_bal.predict(X_val)
lgb_test_proba = lgb_bal.predict(X_test)

# Diagnóstico rápido
print("\n  Probabilidades teste (media):")
for i, cn in enumerate(le_target.classes_):
    p = lgb_test_proba[:, i]
    print(f"    {cn:<12} mean={p.mean():.4f} max={p.max():.4f}")

# argmax
evaluate_model("LightGBM (argmax)", y_test,
               lgb_test_proba.argmax(axis=1), lgb_test_proba)

# threshold
print("\n  Otimizando thresholds:")
t_lgb = optimize_thresholds_ovr(y_val.values, lgb_val_proba, le_target.classes_)
lgb_pred = predict_with_thresholds(lgb_test_proba, t_lgb, verde_idx)
evaluate_model("LightGBM (threshold)", y_test, lgb_pred, lgb_test_proba)

# --- 7.2 LightGBM custo ---
print("\n[7.2] LightGBM (custo de negócio)...")
lgb_params_c = lgb_params.copy()
lgb_params_c['is_unbalance'] = False

dtrain_c = lgb.Dataset(X_train, label=y_train, weight=sample_w_custo)
dval_c   = lgb.Dataset(X_val, label=y_val, reference=dtrain_c)

lgb_custo = lgb.train(
    lgb_params_c, dtrain_c, num_boost_round=3000,
    valid_sets=[dtrain_c, dval_c],
    callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)],
)

lgb_c_val  = lgb_custo.predict(X_val)
lgb_c_test = lgb_custo.predict(X_test)

print("  Otimizando thresholds:")
t_lgb_c = optimize_thresholds_ovr(y_val.values, lgb_c_val, le_target.classes_)
lgb_c_pred = predict_with_thresholds(lgb_c_test, t_lgb_c, verde_idx)
evaluate_model("LightGBM custo (threshold)", y_test, lgb_c_pred, lgb_c_test)

# --- 7.3 XGBoost ---
print("\n[7.3] XGBoost (balanced)...")
xgb_params = {
    'objective': 'multi:softprob', 'num_class': n_classes,
    'eval_metric': 'mlogloss', 'max_depth': 7, 'learning_rate': 0.03,
    'subsample': 0.7, 'colsample_bytree': 0.7, 'min_child_weight': 1,
    'gamma': 0.1, 'reg_alpha': 0.05, 'reg_lambda': 0.5,
    'tree_method': 'hist', 'random_state': 42, 'verbosity': 0,
}

dxgb_t = xgb.DMatrix(X_train, label=y_train, weight=sample_w_balanced)
dxgb_v = xgb.DMatrix(X_val, label=y_val)
dxgb_e = xgb.DMatrix(X_test, label=y_test)

xgb_bal = xgb.train(
    xgb_params, dxgb_t, num_boost_round=3000,
    evals=[(dxgb_t, 'train'), (dxgb_v, 'val')],
    early_stopping_rounds=80, verbose_eval=False,
)

xgb_v_proba = xgb_bal.predict(dxgb_v)
xgb_t_proba = xgb_bal.predict(dxgb_e)

print("  Otimizando thresholds:")
t_xgb = optimize_thresholds_ovr(y_val.values, xgb_v_proba, le_target.classes_)
xgb_pred = predict_with_thresholds(xgb_t_proba, t_xgb, verde_idx)
evaluate_model("XGBoost (threshold)", y_test, xgb_pred, xgb_t_proba)

# --- 7.4 Hierárquico (FIX: sem double-weighting) ---
print("\n[7.4] Hierárquico (2 estágios) — CORRIGIDO")

y_train_bin = (y_train != verde_idx).astype(int)
y_val_bin   = (y_val   != verde_idx).astype(int)
y_test_bin  = (y_test  != verde_idx).astype(int)

# SEM sample weights externo — só scale_pos_weight
lgb_bin_params = {
    'objective': 'binary',
    'metric': ['binary_logloss', 'auc'],
    'num_leaves': 63,
    'learning_rate': 0.03,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'bagging_freq': 5,
    'min_child_samples': 5,
    'scale_pos_weight': (len(y_train_bin) - y_train_bin.sum()) / y_train_bin.sum(),
    'verbose': -1,
    'random_state': 42,
}

# SEM weight no Dataset (fix do double-weighting)
dt_bin = lgb.Dataset(X_train, label=y_train_bin)
dv_bin = lgb.Dataset(X_val, label=y_val_bin, reference=dt_bin)

s1 = lgb.train(
    lgb_bin_params, dt_bin, num_boost_round=3000,
    valid_sets=[dt_bin, dv_bin],
    callbacks=[lgb.early_stopping(80), lgb.log_evaluation(0)],
)
print(f"  Estágio 1: {s1.best_iteration} iterações")

# AUC do estágio 1
s1_val_p = s1.predict(X_val)
s1_auc = roc_auc_score(y_val_bin, s1_val_p)
print(f"  Estágio 1 AUC (val): {s1_auc:.4f}")

# Threshold binário
best_t1, best_f1 = 0.5, 0
for t in np.arange(0.01, 0.99, 0.005):
    p = (s1_val_p >= t).astype(int)
    f = f1_score(y_val_bin, p, average='macro', zero_division=0)
    if f > best_f1:
        best_f1 = f
        best_t1 = t
print(f"  Threshold: {best_t1:.3f} (F1M val: {best_f1:.4f})")

s1_test_p = s1.predict(X_test)
s1_pred = (s1_test_p >= best_t1).astype(int)
rec_nv = s1_pred[y_test_bin == 1].mean() if y_test_bin.sum() > 0 else 0
print(f"  Teste: {s1_pred.sum()} Não-Verde / {y_test_bin.sum()} reais (Recall: {rec_nv:.3f})")

# Estágio 2
nv_classes = [c for c in range(n_classes) if c != verde_idx]
nv_remap = {old: new for new, old in enumerate(nv_classes)}
nv_inv = {v: k for k, v in nv_remap.items()}

X_nv = X_train[y_train != verde_idx]
y_nv = y_train[y_train != verde_idx].map(nv_remap)
nv_counts = np.bincount(y_nv, minlength=len(nv_classes))
nv_w = len(y_nv) / (len(nv_classes) * nv_counts + 1)
nv_sw = y_nv.map(dict(enumerate(nv_w))).values

X_nv_v = X_val[y_val != verde_idx]
y_nv_v = y_val[y_val != verde_idx].map(nv_remap)

s2_params = {
    'objective': 'multiclass', 'num_class': len(nv_classes),
    'metric': 'multi_logloss', 'num_leaves': 31, 'learning_rate': 0.03,
    'feature_fraction': 0.7, 'bagging_fraction': 0.7, 'bagging_freq': 5,
    'min_child_samples': 3, 'is_unbalance': True,
    'verbose': -1, 'random_state': 42,
}

dt_nv = lgb.Dataset(X_nv, label=y_nv, weight=nv_sw)
dv_nv = lgb.Dataset(X_nv_v, label=y_nv_v, reference=dt_nv)

s2 = lgb.train(
    s2_params, dt_nv, num_boost_round=2000,
    valid_sets=[dt_nv, dv_nv],
    callbacks=[lgb.early_stopping(50), lgb.log_evaluation(0)],
)
print(f"  Estágio 2: {s2.best_iteration} iterações")

# Combinar
hier_pred = np.full(len(y_test), verde_idx)
nv_mask = s1_pred == 1
if nv_mask.sum() > 0:
    s2_p = s2.predict(X_test[nv_mask])
    hier_pred[nv_mask] = np.array([nv_inv[p] for p in s2_p.argmax(axis=1)])

evaluate_model("Hierárquico (2 estágios)", y_test, hier_pred)

# ============================================================================
# 8. COMPARAÇÃO CONSOLIDADA
# ============================================================================
print("\n" + "=" * 70)
print("8. COMPARAÇÃO CONSOLIDADA")
print("=" * 70)

rows = {}
for name, r in results.items():
    row = {
        'Acurácia': r['accuracy'],
        'F1 Macro': r['f1_macro'],
        'AUC-ROC': r.get('auc_ovr'),
    }
    for i, cn in enumerate(le_target.classes_):
        row[f'F1 {cn}'] = r['f1_per_class'][i]
    for i, cn in enumerate(le_target.classes_):
        if cn != 'VERDE':
            row[f'Rec {cn}'] = r['recall_per_class'][i]
    rows[name] = row

comparison = pd.DataFrame(rows).T.round(4)
print(f"\n{comparison.to_string()}")

best_name = comparison['F1 Macro'].idxmax()
print(f"\n★ Melhor por F1 Macro: {best_name} ({comparison.loc[best_name, 'F1 Macro']:.4f})")

# Melhor modelo com recall de VERMELHO > 0 (mais útil para negócio)
verm_col = f'Rec VERMELHO'
threshold_models = comparison[comparison[verm_col] > 0]
if not threshold_models.empty:
    best_useful = threshold_models['F1 Macro'].idxmax()
    print(f"★ Melhor com Recall VERM > 0: {best_useful} "
          f"(F1M={threshold_models.loc[best_useful, 'F1 Macro']:.4f}, "
          f"Rec.VERM={threshold_models.loc[best_useful, verm_col]:.4f})")

# ============================================================================
# 9. MATRIZES DE CONFUSÃO
# ============================================================================
print("\n" + "=" * 70)
print("9. MATRIZES DE CONFUSÃO")
print("=" * 70)

# Só modelos com threshold (exclui argmax e baseline ingênuo)
show_models = [n for n in results if 'threshold' in n.lower() or 'hierárquico' in n.lower()]
if not show_models:
    show_models = list(results.keys())[:4]

ncols = min(3, len(show_models))
nrows = max(1, (len(show_models) + ncols - 1) // ncols)
fig, axes = plt.subplots(nrows, ncols, figsize=(7*ncols, 6*nrows))
axes = np.atleast_1d(axes).ravel()

for idx, name in enumerate(show_models):
    ax = axes[idx]
    cm = confusion_matrix(y_test, results[name]['y_pred'], labels=range(n_classes))
    cm_norm = cm.astype(float) / np.maximum(cm.sum(axis=1, keepdims=True), 1)
    annot = np.empty_like(cm, dtype=object)
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            annot[i, j] = f"{cm[i, j]}\n({cm_norm[i, j]*100:.1f}%)"
    sns.heatmap(cm_norm, annot=annot, fmt='', cmap='Blues', ax=ax,
                xticklabels=le_target.classes_, yticklabels=le_target.classes_,
                vmin=0, vmax=1, linewidths=0.5)
    f1m = results[name]['f1_macro']
    auc = results[name].get('auc_ovr')
    auc_s = f" AUC={auc:.3f}" if auc else ""
    ax.set_title(f'{name}\nF1M={f1m:.4f}{auc_s}', fontsize=10, fontweight='bold')
    ax.set_ylabel('Real'); ax.set_xlabel('Predito')

for ax in axes[len(show_models):]:
    ax.axis('off')

plt.tight_layout()
plt.savefig('confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()

# ============================================================================
# 10. FEATURE IMPORTANCE
# ============================================================================
print("\n" + "=" * 70)
print("10. FEATURE IMPORTANCE")
print("=" * 70)

lgb_imp = pd.Series(
    lgb_bal.feature_importance(importance_type='gain'), index=feature_cols
).sort_values(ascending=False).head(25)

fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x=lgb_imp.values, y=lgb_imp.index, palette='viridis_r', edgecolor='k', ax=ax)
ax.set_title('LightGBM — Feature Importance (Gain) Top 25', fontsize=14, fontweight='bold')
ax.set_xlabel('Gain')
sns.despine(ax=ax)
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nTop 15:")
for i, (f, v) in enumerate(lgb_imp.head(15).items(), 1):
    print(f"  {i:2d}. {f:<35} gain = {v:,.0f}")

# ============================================================================
# 11. ROBUSTEZ SAZONAL
# ============================================================================
print("\n" + "=" * 70)
print("11. ROBUSTEZ SAZONAL")
print("=" * 70)

# Usar o melhor modelo com threshold (o mais útil)
eval_name = best_useful if 'best_useful' in dir() else best_name

df_eval = pd.DataFrame({
    'y_true': y_test.values,
    'y_pred': results[eval_name]['y_pred'],
    'yearmonth': df.loc[y_test.index, 'yearmonth'].values,
})

verm_idx_v = le_target.transform(['VERMELHO'])[0]
print(f"\n{eval_name} por mês:\n")
print(f"  {'Mês':<10} {'N':>6} {'Acc':>8} {'F1M':>8} {'Rec.Verm':>10}")
print(f"  {'─'*44}")
for ym in sorted(df_eval['yearmonth'].unique()):
    s = df_eval[df_eval['yearmonth'] == ym]
    acc = accuracy_score(s['y_true'], s['y_pred'])
    f1m = f1_score(s['y_true'], s['y_pred'], average='macro', zero_division=0)
    mv = s['y_true'] == verm_idx_v
    rv = (s.loc[mv, 'y_pred'] == verm_idx_v).mean() if mv.sum() > 0 else float('nan')
    print(f"  {pd.Timestamp(ym).strftime('%Y-%m'):<10} {len(s):>6} {acc:>8.4f} {f1m:>8.4f} {rv:>10.3f}")

# ============================================================================
# 12. SALVAR
# ============================================================================
print("\n" + "=" * 70)
print("12. ARTEFATOS")
print("=" * 70)

lgb_bal.save_model('lgb_balanced.txt')
lgb_custo.save_model('lgb_custo_negocio.txt')
xgb_bal.save_model('xgb_balanced.json')

metadata = {
    'feature_cols': feature_cols,
    'target_classes': list(le_target.classes_),
    'best_model': best_name,
    'thresholds_lgb': {le_target.classes_[i]: float(t_lgb[i]) for i in range(n_classes)},
}
with open('model_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2, ensure_ascii=False)

print("  ✓ Salvos: modelos, metadata, plots.")
print("\n✓ Pipeline v4 completo.")
```
